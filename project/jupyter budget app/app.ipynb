```python
import pandas as pd
from ipywidgets import widgets, VBox ,Output
from IPython.display import display


try:
    transaction = pd.read_csv("transactions.csv")
except FileNotFoundError:
    transaction = pd.DataFrame(columns=["date", "type", "category", "amount", "notes"])


date_picker = widgets.DatePicker(description="Date")
type_dd = widgets.Dropdown(options=["expense", "income"], description="Type")
category_text = widgets.Text(value="", description="Category")
amount_text = widgets.FloatText(value=0.0, description="Amount")
notes_text = widgets.Text(value="", description="Notes")
add_button = widgets.Button(description="Add Transaction", button_style="success")
output = Output()

def add_transaction(b):
    global transaction
    new_row = {
        "date": date_picker.value,
        "type": type_dd.value,
        "category": category_text.value,
        "amount": amount_text.value,
        "notes": notes_text.value
    }
    
    transaction = pd.concat([transaction, pd.DataFrame([new_row])], ignore_index=True)
    
    transaction.to_csv("transactions.csv", index=False)
    with output:
        output.clear_output()
        print("Transaction saved . Last 5 entries:")
        display(transaction.tail(5))
    


add_button.on_click(add_transaction)


display(VBox([date_picker, type_dd, category_text, amount_text, notes_text, add_button]))
```
